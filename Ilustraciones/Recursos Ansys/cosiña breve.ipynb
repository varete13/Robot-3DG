```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from matplotlib import ticker
```

```python
df=pd.read_excel('Modos Propios.xlsx')
```

```python
n=df['Mode']
y=df['Frequency [Hz]']
```

```python
plt.style.use('fivethirtyeight')
```

```python
fig,ax = plt.subplots(figsize=(15,5))
hbar=ax.barh(n,y)
ax.grid(False)
ax.bar_label(hbar)
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:.0f}'))
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,}'))
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x,pos: format(x/1000,'1.0f')+' KHz'))
ax.yaxis.set_visible(False)
ax.grid(True,which='both',axis='x')

fig.suptitle('Modos Propios')
# ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x} Hz'))
fig.autolayout:True
plt.savefig('Modos Propios.png',dpi=300)


# plt.table(cellText=[y],rowLabels=['Modos'],loc='bottom')
```
